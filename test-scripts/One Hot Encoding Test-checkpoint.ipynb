# Results with One Hot Encoding

```python
#hardcoded vals for testing
features = ["Region", "GDP per capita", "Social support",
                "Healthy life expectancy", "Freedom to make life choices", "Generosity", "Perceptions of corruption"]
target = "Score"
title = "testing testing"
testSize = 0.1
randomState = 3 # for reproducible output
numFolds = 12
```

```python
# create pandas dataframe
import pandas as pd

data = pd.read_csv("/Users/quinnwaiwong/projects/introdatascience/homeworks/hw2-quinnwai/utility/data/happiness.csv")

X = data.loc[:,features]
y = data.loc[:,target]
print(X.shape)
data.head(3)
```

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np

str_indices = []
for i in range(0, len(features)):
    if(isinstance(X.loc[0, features[i]], str)):
        str_indices.append(i)

# if strings columns exist, split dataset into numeric and string data
if (len(str_indices) != 0):
    
    features = np.array(features)
    str_features = features[str_indices]
    X_str = X[str_features]
    
#     print(f'X_str {X_str}')
    
    enc = OneHotEncoder().fit(X_str)
    
#     print(f'enc.categories_ {type(enc.categories_[1])}')
    
    for i in range(len(enc.categories_)):
        print(i)
        if(i == 0):
            columns = enc.categories_[i]
        else:
            columns = np.concatenate((columns, enc.categories_[i]))
    
#     print(f'columns {columns}')
    
    X_enc = pd.DataFrame(enc.transform(X_str).toarray(), columns=columns)
    
#     for col in X_enc.columns:
#         assert sum(X_enc.loc[:,col]) == 1, "not 1"
    
    print(f'X_enc {X_enc.shape} and head {X_enc.head(3)}')

    X_num = X.drop(str_features, axis=1)

    X = pd.concat([X_num, X_enc], axis=1)

print(X.head(3))
```

## Mario time

```python
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression

# initialize KFolds
kf = KFold(n_splits=numFolds, shuffle=True, random_state=randomState)

score = 0
##### cross validation and model evaluation #####
for fold, (train_idx, test_idx) in enumerate(kf.split(X)):
    ## pre-processing phase
    #define train and test 
    X_train = X.loc[train_idx,:]
    X_test = X.loc[test_idx,:]
    y_train = y.loc[train_idx]
    y_test = y.loc[test_idx]

    ## use k-NN to predict and get r squared
    # create and fit model
    model = LinearRegression().fit(X_train, y_train)

    # predict using new model and compute r_squared value
    y_pred = model.predict(X_test)
    r_squared = model.score(X_test, y_test)

    # sum
    score += r_squared
    
    print(f'r squared for fold {fold} is {r_squared}')

avg_score = score/(fold+1)

print(f'average score is {avg_score}')
```

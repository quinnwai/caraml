# Visualization

```python
## calculation-specific imports
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression

## graph-specific imports/settings
import matplotlib
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('/Users/quinnwaiwong/projects/introdatascience/homeworks/hw2-quinnwai/utility/data/happiness.csv')
```

```python
data.columns
```

```python
target = 1
y = data.iloc[:,target]
allFeatures = data.columns.drop(data.columns[target])
```

```python
i = 0
for i in range(len(allFeatures)):
    feature = allFeatures[i]
    X = data.loc[:,feature]
    if np.issubdtype(X.dtype, np.integer) or np.issubdtype(X.dtype, np.floating):
        plt.scatter(X, y)
        plt.title(f'{target} vs {feature}')
        plt.xlabel(feature)
        plt.ylabel(target)
        plt.show()
```
